```python
import re
text = []

with open('_chat.txt', 'r', encoding="utf8") as f:
        for line in f:
            line = re.sub(r'\[.*\]|Here|song|open|spotify|track|call|voice|image|omitted|kuch|nahi|Mere|video|apni|dekh|Okay|Haan|baat|sahi|tere|mere|Nahi|baad|hoga|bhi|teri|baar|know|mujhe|kerna|tera|pata|tujhe|Okay|aisa|kaafi|abhi|liye|Tune|acha|mera', r'', line)
            text.append(line)
        
```

```python
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)

text = list(map(deEmojify, text))
```

```python
#def deTime(text):
    #return re.sub(r'\[.*\]', r'', text)
chat = map(lambda x: re.sub(r'.*:', r'', x), text)
```
